### **Visualize the iris database**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

#Looking at the data for variance
iris=datasets.load_iris()
df = pd.read_csv('Iris.csv')

df.describe()
```

### Most of the data seems pretty normal except for PetalWidthCm when min in 0.1 and max is 2.5

```python
#plotting a histogram fit PetalWidthCm to see if it is a small subset
df['PetalWidthCm'].plot.hist()
plt.show()
```

```python
#That seems about one third of the data is between 0-0.5
#Lets try plotting different types
species=list(set(df['Species']))
#fig, axes = plt.subplots(1, 3 ,figsize=(15,6))
df.hist(column='PetalWidthCm',by='Species',)
```

  ### We can clearly see the small values are limited to iris-setosa. So maybe thats a property of the flower

```python
#Next up we use seaborn to see how features vary across the targets
sns.pairplot(df, hue='Species')
```

### Most of the cross scatter of the features across all targets are fairly separated already indicating that is a easy to train dataset.
